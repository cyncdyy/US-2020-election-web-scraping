```python
import plotly
import pandas as pd
import plotly.express as px

# Clean the data
df = pd.read_csv('house.csv')
df['Vote'] = df['Vote'].replace({'/': '0', ',': ''}, regex=True).astype(int)
df['Pct%'] = df['Pct%'].replace({'/': '0', '%': ''}, regex=True).astype(float)
df['District'] = df['District'].replace({'/' : 'Only 1 District'}, regex=True)
df['Winner'] = 'no'
groupby_state_district = df.groupby(['State', 'District'])
for state, district in groupby_state_district:
    first_row_index = district.index[0]
    df.loc[first_row_index, 'Winner'] = 'yes'
```

```python
state_unique = list(df['State'].unique())
state_code_unique = list(df['State Code'].unique())
Total_title_seats = 0
Republican_title_seats = 0
Democratic_title_seats = 0

for j in range(len(state_unique)):
    for i in range(len(df)):
        if df['State'][i] == state_unique[j]:
            if df['Winner'][i] == "yes":
                Total_title_seats +=1
                if df['Party'][i] == 'Republican':
                    Republican_title_seats +=1
                else:
                    Democratic_title_seats +=1          
```

```python
state_unique = list(df['State'].unique())
state_code_unique = list(df['State Code'].unique())
Total_seats = []
Republican_seats = 0
Democratic_seats = 0
Won_seats = []
Won_party = []
Republican_votes = []
Democratic_votes = []
Libertarian_votes = []
Green_votes = []
Independent_votes = []
Won_percent = []
for j in range(len(state_unique)):
    T_Win_sum = 0
    R_Win_sum = 0
    D_Win_sum = 0
    R_Total_sum = 0
    D_Total_sum = 0
    L_Total_sum = 0
    G_Total_sum = 0
    I_Total_sum = 0
    for i in range(len(df)):
        if df['State'][i] == state_unique[j]:
            if df['Winner'][i] == "yes":
                T_Win_sum +=1
                if df['Party'][i] == 'Republican':
                    R_Win_sum +=1
                else:
                    D_Win_sum +=1            
            if df['Party'][i] == 'Republican':
                R_Total_sum += df['Vote'][i]
            elif df['Party'][i] == 'Democratic':
                D_Total_sum += df['Vote'][i]
            elif df['Party'][i] == 'Libertarian Party':
                L_Total_sum += df['Vote'][i]
            elif df['Party'][i] == 'Green Party':
                G_Total_sum += df['Vote'][i]
            elif df['Party'][i] == 'Ind.':
                I_Total_sum += df['Vote'][i]
    Total_seats.append(T_Win_sum)
    Republican_seats += R_Win_sum
    Democratic_seats += D_Win_sum
    if R_Win_sum > D_Win_sum:
        Won_seats.append(R_Win_sum)
        Won_party.append('Republican')
    elif R_Win_sum < D_Win_sum:
        Won_seats.append(D_Win_sum)
        Won_party.append('Democratic')   
    else:
        Won_seats.append(R_Win_sum) 
        Won_party.append('Draw')  
    Republican_votes.append(R_Total_sum)
    Democratic_votes.append(D_Total_sum)
    Libertarian_votes.append(L_Total_sum)
    Green_votes.append(G_Total_sum)
    Independent_votes.append(I_Total_sum)
for i in range(len(state_unique)):
    if Won_party[i] == 'Republican':
        Won_percent.append(100*Won_seats[i]/Total_seats[i])       
    else:
        try:
            Won_percent.append(100*(Total_seats[i]-Won_seats[i])/Total_seats[i])
        except:
            print('Total', Total_seats[i], 'Won', Won_seats[i])
```

```python
# Combine the dataset to a dataframe
data = {
    'State': state_unique,
    'State Code': state_code_unique,
    'Won Party': Won_party,
    'Won Seats': Won_seats,
    'Total Seats': Total_seats,   
    'Republican Votes': Republican_votes,
    'Democratic Votes': Democratic_votes,
    'Libertarian Party Votes': Libertarian_votes,
    'Green': Green_votes,
    'Independent': Independent_votes,
    'Rep Won Seats %': Won_percent,
}
dff = pd.DataFrame(data)
dff
```

```python
# final

from jupyter_dash import JupyterDash
from dash import html, dcc
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import plotly.express as px
import pandas as pd
from urllib.request import urlopen
import json
import plotly.graph_objects as go

# Load the GeoJSON data for the map
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    state = json.load(response)

# Create the JupyterDash app
app = JupyterDash(__name__)

# Define the hoverlist for map hover data
hoverlist = ['State', 'Won Party', 'Won Seats', 'Total Seats', 'Republican Votes', 'Democratic Votes', 'Libertarian Party Votes']

# Create the choropleth map figure
fig = px.choropleth(dff, locationmode="USA-states", geojson=state, scope="usa",
                    locations='State Code', color="Rep Won Seats %",
                    color_continuous_scale="Rdbu_r",
                    range_color=(0, 100),
                    hover_name='State',
                    hover_data=hoverlist)
parties_dropdown = dcc.Dropdown(
    id='parties-dropdown',
    options=[{'label': party.replace(' Votes', ''), 'value': party} for party in dff.columns[5:10]],
    multi=True,
    placeholder='Select parties'
)

# Set up the app layout
app.layout = html.Div([
    html.H1("US House Election 2022"),
    html.Hr(),
    html.Div([
        html.H2(f"Total seats: {Total_title_seats}"+"\n" + '|' +"\n"+
                          f"Republican won: {Republican_title_seats}"+"\n" + '|' +"\n"+
                          f"Democratic won: {Democratic_title_seats}",style={"flex":"1"}),
        html.H3(f"Winning Party:{title_winning_party}",style={"flex":"1","text-align":"center"})
    ],style={"display":"flex","align-items":"center","justify-content":"space between"}),
    html.Div([dcc.Graph(id='map', figure=fig, hoverData={'points': [{'customdata': ['Alabama']}]})],
             style={'width':'65%', 'float':'left', 'display':'inline-block'}),
    html.Div([dcc.Graph(id='bar-chart-state', style={'height': '300px'})],
             style={'width': '35%', 'float': 'right', 'display': 'inline-block'}),
    html.Div([dcc.Graph(id='bar-chart-district', style={'height': '300px'})],
             style={'width': '35%', 'float': 'right', 'display': 'inline-block'}),
    html.Div([
    html.Div([parties_dropdown], style={'width': '100%', 'display': 'inline-block', 'padding': '0 20'}),
    html.Div(id='pie-chart-container')  # pie-chart-container
])
    ])

# Callback to update the bar chart for the winning party in each state
@app.callback(
    Output('bar-chart-state', 'figure'),
    Input('map', 'hoverData')
)
def update_bar_chart_state(hover_Data):
    if not hover_Data:
        raise PreventUpdate  # Don't update the plot if no hover data is available

    selected_state = hover_Data['points'][0]['customdata'][0]

    # Filter the data for the selected state
    state_data = df[df['State'] == selected_state]
    party_votes = state_data.groupby('Party')['Vote'].sum().reset_index()

    # Create the bar chart
    fig_bar = px.bar(party_votes, x='Party', y='Vote', color='Party',
                     color_discrete_map={"Republican": "red", "Democratic": "blue", "Libertarian": "pink"},
                     labels={'Party': 'Parties', 'Vote': 'Votes'},
                     title=f'States:{selected_state}')

    return fig_bar

# Callback to update the bar chart for the winning party in each district
@app.callback(
    Output('bar-chart-district', 'figure'),
    Input('map', 'hoverData')
)
def update_bar_chart_district(hover_Data):
    if not hover_Data:
        raise PreventUpdate  # Don't update the plot if no hover data is available

    selected_state = hover_Data['points'][0]['customdata'][0]

    # Filter the data for the selected state
    state_data = df[df['State'] == selected_state]
    district_party_votes = state_data.groupby(['State', 'District', 'Party'])['Vote'].sum().reset_index()

    # Filter for the winning party in each district
    winning_party_votes = district_party_votes.groupby(['District'])['Vote'].max().reset_index()
    winning_party_votes = winning_party_votes.merge(district_party_votes, on=['District', 'Vote'], how='left')
    district_count = len(winning_party_votes['District'].unique())
    # Create the bar chart
    fig_bar = px.bar(winning_party_votes, x='District', y='Vote', color='Party',
                     color_discrete_map={"Republican": "red", "Democratic": "blue", "Libertarian": "pink"},
                     labels={'District': 'District', 'Vote': 'Votes'},
                     title=f"No of Districts: {district_count}")

    return fig_bar


# additional feature open callback

@app.callback(
    Output('pie-chart-container', 'children'),
    [Input('parties-dropdown', 'value')]
)
def update_pie_charts(selected_parties):
    if not selected_parties:
        raise PreventUpdate  # don't update the plot if no parties are selected

    pie_charts = []
    for party in selected_parties:
        party_votes = dff[party]
        fig = go.Figure(data=[go.Pie(
            labels=dff['State'],
            values=party_votes,
            hoverinfo='label+value+percent',
            textinfo='none',
            name=party,
            title={'text': f"Vote Distribution of {party.replace(' Votes', '')}", 'position': 'middle center'}
        )])
        pie_charts.append(dcc.Graph(figure=fig))

    return pie_charts



if __name__ == '__main__':
    app.run_server(port=8058, debug=True)
```
